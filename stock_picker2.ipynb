```python
import os
import sys
from datetime import datetime

import gradio as gr
from google.colab import userdata

# API Keys
os.environ["OPENAI_API_KEY"] = userdata.get("OPENAI_API_KEY")
os.environ["SERPER_API_KEY"] = userdata.get("SERPER_API_KEY")

# Project Path
PROJECT_ROOT = "/content/AgenticAI-Lab/stock_picker"
sys.path.insert(0, f"{PROJECT_ROOT}/src")


from stock_picker.crew import StockPicker


def stock_picker(sector):

    if not sector.strip():
        return "Please enter a sector"

    try:
        inputs = {
            "sector": sector,
            "current_date": str(datetime.now().date())
        }

        result = stock_picker().crew().kickoff(inputs=inputs)

        return str(result)

    except Exception as e:
        return f"Error:\n\n{e}"
```

**Step 7: Run the code with gradio UI**

```python
import gradio as gr
from datetime import datetime

from stock_picker.crew import StockPicker


def run_stock_picker(sector):
    """Run the CrewAI Stock Picker Agent."""

    sector = (sector or "").strip()

    if not sector:
        return "Please enter a stock market sector."

    try:
        inputs = {
            "sector": sector,
            "current_date": str(datetime.now().date())
        }

        # Create and execute the CrewAI StockPicker crew
        result = StockPicker().crew().kickoff(inputs=inputs)

        return str(result)

    except Exception as e:
        return f"Error while running the Stock Picker Agent:\n\n{e}"


demo = gr.Interface(
    fn=run_stock_picker,

    inputs=gr.Textbox(
        label="Stock Market Sector",
        placeholder="e.g. Technology, Healthcare, AI, Banking"
    ),

    outputs=gr.Textbox(
        label="Stock Research Report",
        lines=30
    ),

    title="📈 CrewAI Stock Picker Agent",

    description="""
Enter a stock market sector.

The CrewAI Stock Picker Agent will:

• Research recent sector developments
• Identify leading public companies
• Compare investment opportunities
• Analyze growth potential and risks
• Generate a detailed stock research report
""",

    examples=[
        ["Technology"],
        ["Artificial Intelligence"],
        ["Healthcare"],
        ["Cybersecurity"],
        ["Renewable Energy"]
    ]
)

demo.launch(
    share=True,
    debug=True
)
```
